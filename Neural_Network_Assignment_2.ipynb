```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
img_data = pd.read_csv('Q8_image.csv', header=None).values
kernel = np.array([
    [0.053, 0.110, 0.054],
    [0.111, 0.225, 0.111],
    [0.054, 0.110, 0.053]
])
filtered_image = convolve2d(image_data, kernel, mode='same', boundary='fill', fillvalue=0)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_data, cmap='gray')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title("Filtered Image")
plt.imshow(filtered_image, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()
```
