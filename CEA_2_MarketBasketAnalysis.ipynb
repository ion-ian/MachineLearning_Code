```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = []
with open('Market_Basket_Optimisation.csv') as file:
    reader = csv.reader(file, delimiter=',')
    for row in reader:
        dataset += [row]
```

```python
dataset[0:10]
```

```python
len(dataset)
```

```python
te = TransactionEncoder()
x = te.fit_transform(dataset)
```

```python
x
```

```python
df = pd.DataFrame(x, columns=te.columns_)
```

```python
te.columns_
```

```python
len(te.columns_)
```

```python
df.head()
```

```python
# Finding Frequent Itemsets
freq_itemset = apriori(df, min_support=0.01, use_colnames=True)
```

```python
freq_itemset
```

```python
num_itemsets = len(freq_itemset)
rules = association_rules(freq_itemset, metric="confidence", min_threshold=0.25, num_itemsets=num_itemsets)
```

```python
rules
```

```python
rules = rules[['antecedents','consequents','support','confidence']]
```
